# Crop Yield Dataset - Data Preprocessing



```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('crop_yield1.csv')
df.head()
```

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
for col in cat_cols:
    df[col] = df[col].str.strip()
```

```python
for col in cat_cols:
    df[col] = df[col].str.lower()
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
columns_to_check = ['Soil_Type', 'Region', 'Season', 'Crop_Type', 'Irrigation_Type']

for col in columns_to_check:
    print(f"Unique values in '{col}':\n", df[col].value_counts(), "\n")
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding for the columns with repeated values
le = LabelEncoder()
for col in columns_to_check:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
for col in columns_to_check:
    print(f"\n{col} Mapping:")
    print(dict(zip(le.classes_, range(len(le.classes_)))))
```

```python
df.to_csv("crop_yield_cleaned.csv", index=False)
```
