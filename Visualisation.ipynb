```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("c:\\Users\\TARUN KUMAR\\Desktop\\Datasets\\crop-yield.csv") 


df.head()
df.info()
df.describe()
```

1 Scatter Plot – Rainfall vs Yield

```python
df.plot(x="Rainfall", y="Crop_Yield_ton_per_hectare", kind="scatter", title="Rainfall vs Yield")
plt.show()
```

2 Pie Chart – Crop Type Distribution


```python
df["Crop_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%", title="Crop Type Distribution")
plt.ylabel("")
plt.show()
```

3 Line Plot – Humidity vs Crop Yield

```python
df.plot(x="Humidity", y="Crop_Yield_ton_per_hectare", kind="hist", title="Humidity vs Crop Yield")
plt.show()
```

4 Bar Plot – Average Yield by Region

```python
df.groupby("Region")["Crop_Yield_ton_per_hectare"].mean().plot(kind="bar", title="Average Yield by Region")
plt.show()
```

6 Bar Plot – Average Yield by Crop Type

```python
df.groupby("Crop_Type")["Crop_Yield_ton_per_hectare"].mean().plot(kind="bar", title="Average Yield by Crop Type")
plt.show()
```

7 Bar Plot – Average Yield by Season

```python
df.groupby("Season")["Crop_Yield_ton_per_hectare"].mean().plot(kind="bar", title="Average Yield by Season")
plt.show()
```

8 Horizontal Bar Plot – Irrigation Type

```python
df.groupby("Irrigation_Type")["Crop_Yield_ton_per_hectare"].mean().plot(kind="barh", title="Yield by Irrigation Type")
plt.show()
```
